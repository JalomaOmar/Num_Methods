```python
#NECESARIO PARA LAS IMPLEMENTACIONES
import numpy

from numpy import float32 as f32 

def machineEpsilon(func=f32):#Esta función nos permite saber el machine epsilon
    machine_epsilon = func(1)
    while func(1)+func(machine_epsilon) != func(1):
        machine_epsilon_last = machine_epsilon
        machine_epsilon = func(machine_epsilon) / func(2)
    return machine_epsilon_last

feps=machineEpsilon(float) #feps es el machine epsilon
print(feps)

#----------------------------------------------------------------------------------------------------------------------------------------


#BISECTION

def Bisection(f, a, b, eps): #f debe ser una función lambda, a y b son son los extremos del intervalo y eps el epsilon 
    x=a+((b-a)/2)
    j=0
    while(b-a>eps and abs(f(x))>eps and j<numpy.log2(1/eps)):
      if(f(a)*f(x)<0):
         b=x
      else:
         a=x
      j+=1 
    return([x, f(x), (abs((b-a))/max(1, x)), j])     

f= lambda x: x**2-1
print(Bisection(f, 0.98,1.34, feps))


#---------------------------------------------------------------------------------------------------------------------------------------------

#NEWTON-RAPHSON

def NewtonR(f, fp, x0, eps): #f es la función, fp la derivada, x0 el valor inicial y eps nuestro epsilon
    x=x0
    xf=x0+1
    j=0
    while( abs(xf-x)>eps and abs(f(x))>eps and (abs((xf-x))/max(1, xf)) and j<numpy.log2(1/eps)):
      if fp(x)==0:
        break
      xf=x 
      x=x-(f(x)/fp(x))
      j+=1
    return([x, f(x), (abs((xf-x))/max(1, xf)), j])

f= lambda x:x**2-1
fp= lambda x:x*2

print(NewtonR(f,fp, 1.2, feps))
         
#-----------------------------------------------------------------------------------------------------------------------------------------

#SECANTE

def Secante(f, x0, eps):
    x=x0
    xf=x0+1
    j=0
    while( abs(xf-x)>eps and abs(f(x))>eps and (abs((xf-x))/max(1, xf)) and j<numpy.log2(1/eps)):
      if (f(x)-f(xf))==0:
        break
      xr=xf
      xf=x
      x=x-f(x)*( (x-xr )  /  (f(x)-f(xr))) #Aquí ocurre el cambio respecto al código de NEWTON
      j+=1
    return([x, f(x), (abs((xf-x))/max(1, xf)), j])


print(Secante(f, 1.5, feps))
print("\n")


#---------------------------------------------------------------------------------------------------------------------------------------------
#PREGUNTA 2

f= lambda x: numpy.exp(x)+2**(-x)+2*numpy.cos(x)-6
fp= lambda x: numpy.exp(x)-numpy.log(2)*2**(-x)-2*numpy.sin(x)


print(Bisection(f,1,2,feps))
print(NewtonR(f,fp, 1.5, feps ))
print(Secante(f, 1.5, feps))

print("\n")
f= lambda x: numpy.log(x-1)+numpy.cos(x-1)
fp= lambda x: (1/(x-1))-numpy.sin(x-1)

print(Bisection(f,1.3,2,feps))
print(NewtonR(f,fp, 1.5, feps ))
print(Secante(f, 1.5, feps))


print("\n")
#-------------------------------------------------------------------------------------------------------------------------------------------------
#PREGUNTA 3

def f(x):
  if x >=0:
    return(numpy.sqrt(x))
  else:
    return(-numpy.sqrt(-x))

def fp(x):
  if x >=0:
    return(1/(2*numpy.sqrt(x)))
  else:
    return(1/(2*numpy.sqrt(-x)))


g= lambda x: x**(1/3)
gp= lambda x: (1/3)*x**(-2/3)

h= lambda x: (x-1)**(1/2)-numpy.exp(1)+numpy.exp(x)
hp= lambda x: (1/2*((x-1)**(1/2)))+numpy.exp(x)

print(NewtonR(f,fp, 0, feps ))
print(NewtonR(g,gp, 0, feps ))
print(NewtonR(h,hp, 1, feps ))
```
